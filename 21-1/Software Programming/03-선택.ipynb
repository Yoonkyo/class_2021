# 1. 조건문

## 1) 조건문 소개

- 우리가 문제를 해결할 때 어떤 조건에 따라서 두 개 또는 여러 개의 실행 경로 가운데 하나를 선택해야 하는 경우가 종종 있다
- 조건문이란 조건에 따라 결정을 내리는 문장을 의미한다






## 2) if-else 문

- 조건문을 파이썬으로 작성 시 사용
- 조건이 참이면 A를, 참이 아니면 B를 실행하는 것






## 3) 관계연산자

- 관계 연산자(relational operator)는 두 개의 피연산자를 비교하는데 사용된다. 예를 들면 "변수 x가 변수 y보다 큰지"를 따지는데 사용된다




- 관계 수식은 참이나 거짓이라는 값을 생성한다



````python
# if-else문 예제

age = 19
if(age >=  19):
    print("마트에서 주류를 구입할 수 있습니다. ")
else:
    print("조금 기다리세요!")
````

````python
# else는 생략 가능

food = '스테이크'
if food == '스테이크':
    print('내가 제일 좋아하는 음식 !')
    print(10 * food)
````

### 예제
- 학생들의 성적을 받아서 학점을 출력하는 프로그램을 작성하여 실행하여 보자. 성적이 90점 이상이면 A학점, 80점 이상이고 90점 미만이면 B학점, 70점 이상이고 80점 미만이면 C학점과 같이 결정하는 것이다. 

```
성적을 입력하시오: 90
학점 A
```



````python
score = int(input("성적을 입력하시오: "))

if score >= 90:
    print("학점 A")
elif score >= 80:
    print("학점 B")
elif score >= 70 :
    print("학점 C")
elif score >= 60 :
    print("학점 D")
else:
    print("학점 F");
````

## 2) 중첩 if-else 문

- 우리는 if-else 문장 안에 다른 if-else 문장을 넣을 수 있다. 이것을 중첩(nesting)된 if-else문이라고 부른다.





### 예제

- 예를 들어보자. 마트에서 사과가 신선하면 사과를 사기로 한다. 만약 사과가 개당 1000원 미만이면 10개를 산다. 하지만 사과가 개당 1000원 이상이면 5개만 산다. 


````python
appleQuality = input("사과의 상태를 입력하시오: ")
applePrice = int(input("사과 1개의 가격을 입력하시오: "))
if appleQuality == "신선":
    if applePrice < 1000:
        print("10개를 산다")
    else:
        print("5개를 산다")
else:
    print("사과를 사지 않는다.")
````

# 5. 문자열과 숫자

## 1) 문자열과 숫자의 차이

- 숫자 100과 문자열 "100"은 어떤 차이가 있을까? 유사해보이지만 컴퓨터에서는 이것을 상당히 다르게 처리한다. 우리가 조건식을 만들 때도 문자열과 숫자를 구별하여야 한다.  

- 아래 식은 조건식이 거짓이기 때문에 아무것도 출력되지 않는다. 

```python
s = "100"
if (s == 100):
    print("숫자 100입니다. ")
```

## 2) 문자열-> 숫자

- 문자열에서 숫자로 자료형을 변환하기 위해서는 int() 함수를 사용한다.

```python
s = "100"
number=int(s)
if (number==100):
    print("숫자 100입니다. ")
```
