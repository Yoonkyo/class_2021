# 2. 계산

## 1) 수식과 연산자

- 수식(expression)=피연산자들과 연산자의 조합
- 연산자(operator)는 연산을 나타내는 기호
- 피연산자(operand)는 연산의 대상이 되는 것

```python
>>> 3 + 4
7
>>> 3.14 * 5.0 * 5.0
78.5
```



## 2) 산술 연산자
- 덧셈, 뺄셈, 곱셈, 나눗셈, 나머지 연산 



### 나눗셈

- 정수나눗셈은 //, 실수나눗셈은 /을 사용한다

```python
>>> 7 / 4
1.75
>>> 7 // 4 
1
```

### 나머지 계산

- %를 사용하여 나머지를 구한다
```python
>>> 7 % 4  
3
```

- 예제로 초 단위의 시간을 받아서 몇 분 몇 초인지를 계산하여 보자

```python
>>> sec = 1000
>>> min = 1000 // 60
>>> remainder = 1000 % 60
>>> print(min, remainder)
16 40
```

````python
sec = 1000
min = 1000 // 60
remainder = 1000 % 60
print(min, '분', remainder, '초')
````

### 지수 계산

- 지수(power)를 계산하려면 ** 연산자를 사용한다

```python
>>> 2 ** 7 
128
```

- 지수는 곱셈, 나눗셈 연산보다 먼저 계산된다

```python
>>> 3 * 2 ** 3
24
```

- 원리금 계산

```python
>>> a = 1000
>>> r = 0.05
>>> n = 10
>>> a*(1+r)**n
1628.894626777442
```

````python
# 원리금 계산
a = 1000
r = 0.05
n = 10
a*(1+r)**n
````

### 연산자의 우선순위
- 지수, 곱셈/나눗셈, 덧셈/뺄셈

```python
>>> 1 + 2 * 3

>>> 4 - 40 - 3

```

- 괄호를 사용하여 우선순위 변경 가능

```python
>>> 10 + 20 / 2

>>> (10 + 20) / 2
```


## 8) 문자열 응용

### 특수 문자열

- 문자 앞에 \가 붙으면 문자의 특수한 의미를 잃어버린다 

```python
>>> message= 'doesn\'t'  # \를 사용하여 작은따옴표를 출력한다.
>>> print(message)
doesn't
>>>
>>> message="\"Yes,\" he said."
>>> print(message)
"Yes," he said.
>>> 
```

### 문자열의 연결

- +를 사용하여 결합 가능하다

```python
>>> first_name="길동"
>>> last_name="홍"
>>> name = last_name+first_name
>>> print(name)
홍길동
>>> 
```


### 문자열과 정수 간의 변환

- 문자열과 정수는 연결될 수 없으므로 자료형을 변환해야 한다


```python
>>> "Student"+26
...

>>> price = int("259000")
>>> height = float("290.54");
```

### 문자열의 반복

- *을 사용하여 동일한 문자열을 반복할 수 있다

```python

>>> print(line)

>>> message = "Congratulations! "

```

### 문자열의 출력

- %s를 활용하여 원하는 문자열을 출력할 수 있다

```python
>>> price = 10000

상품의 가격은 10000원입니다.

>>> message = "현재 시간은 %s입니다."
>>> time = "12:00pm"
>>> print(message % time)
현재 시간은 12:00pm입니다.
```

```python

>>> 
```

```python
>>> shopping_list[2]='apple'

```
